```
'''
Name:Swastik Pradhan
Email for LaunchEd:swastikpradhan2003@gmail.com
'''
'''
Import the neccesary modules to operate on given Project:
1)Import Numpy for Numerical based operations
2)Import Pandas to work with the Data series or structured data(Excel file)
3)Import Sklearn to get the predefined code for training test split,and for logistic regression
'''

import numpy
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


'''
load the excel file by copying the path of the excel in files of google collab
'''
'''
Content of the file is located in /content/sample_data/Capstone Project(Cleaned).xlsx
'''
df = pd.ExcelFile("/content/drive/MyDrive/Data analytics Capstone project/Capstone Project(Cleaned).xlsx")

# Check available sheet names
print("Available sheets:", df.sheet_names)

# Check for the sheet name if it correct sheet uploaded or not
data = df.parse(df.sheet_names[0])  # For example, load the first sheet
#print(data.head())

'''
feature_cols = [col for col in data.columns if col != 'Fail']  line creates a list of column names
of the data frame except the Target variable 'Fail' or "Not Fail" Column.
'''
'''
X = data[feature_cols] helps in extracting columns from data frame from feature columns and assign to X
'''
'''
Y = data['Fail'] this line helps in extracting the target column from the data frame and assign to Y.
'''
'''
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y    this line helps in spliting the data
    into training and testing data for X and Y variable. Test_size is to determine the size of test data
    which 0.2 in this case
)

random_state is used to make your results reproducible
'''
feature_cols = [col for col in data.columns if col != 'Fail']  # all except target
X = data[feature_cols]
Y = data['Fail']

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=42, stratify=Y
)

#intialize the model Logistic Regression from Sklearn library
'''
max_iter sets an max limit on how many steps
(iterations) it will perform to try to find the best solution by optimizing parameters to
minimize loss function.
'''
'''
Model then is fed with the input Xtrain and corresponding output as Ytrain for the Logistic Regression model
to learn the patterns and relationships in the data.
'''
model=LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)

#In addition just import the confusion matrix to get idea of Accuracy, Precision,Roc Curve etc
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

'''
Y_pred = model.predict(X_test) this line is used to make trained logistic model for each observation
in test case of X_test

Y_proba = model.predict_proba(X_test)[:, 1]  returns a 2-dimensional array where each
row contains the probabilities the model assigns to each class for a given test observation.

'''
Y_pred = model.predict(X_test)
Y_proba = model.predict_proba(X_test)[:, 1]

print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
print("\nClassification Report:\n", classification_report(Y_test, Y_pred))
print("ROC-AUC Score:", roc_auc_score(Y_test, Y_proba))
```
